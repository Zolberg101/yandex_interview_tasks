```python
import urllib2
import pandas
from sets import Set
import re
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook

    
def repl(x): 
    q = re.sub(u' ÑÐ½Ð²', '.01.2016', x, flags=re.U)
    q = re.sub(u' Ñ„ÐµÐ²', '.02.2016', q, flags=re.U)
    q = re.sub(u' Ð¼Ð°Ñ€', '.03.2016', q, flags=re.U)
    q = re.sub(u" Ð°Ð²Ð³", '.08.2015', q, flags=re.U)    
    q = re.sub(u' ÑÐµÐ½', '.09.2015', q, flags=re.U)
    q = re.sub(u' Ð¾ÐºÑ‚', '.10.2015', q, flags=re.U)
    q = re.sub(u' Ð½Ð¾Ñ', '.11.2015', q, flags=re.U)
    q = re.sub(u' Ð´ÐµÐº', '.12.2015', q, flags=re.U)
    return q

def get_liveinternet(url):

    opener = urllib2.build_opener()
    # opener.addheaders = [
    #                         ('Accept-Language', 'en-US,en;q=0.8'), 
    #                         ('Cookie', 'suid=2K8WjN1OJgHR; _ym_uid=1455019082148753305; chbx=guest; lang=en')
    #                     ]
    resp = opener.open(url)
    text = resp.read().decode('utf-8')

    data = text.split("\n")

    for i in xrange(len(data)):
        data[i] = data[i].split(';')

    columns=["date", "yandex","google","mail","rambler", "bing"]

    df = pandas.DataFrame(data[1:],columns=columns)
    df.index = df.date
    df.drop("date", axis=1,inplace=True)
    df.dropna(inplace=True)
    dates = map(lambda x: x.split(" "), df.index.tolist())
    df.index=df.index.map(repl)
    dates = df.index.tolist()
    for i in xrange(len(dates)):
        dates[i] = dates[i].strip('"')
        date = map(int, dates[i].split("."))
        dates[i] = datetime(date[2],date[1], date[0])
    
    df.index = pandas.DatetimeIndex(dates)

    f = plt.figure(figsize=(13, 5), dpi=80)


    for c in columns[1:3]:
        df[c] = df[c].map(int)    
        df[c].plot(legend=True)
        plt.subplots_adjust(left=0.05, bottom=0.1, right=0.87, top=0.93, wspace=0.27, hspace=0.36)

    plt.show()
    
    return dates
                       
                       
dates = get_liveinternet("http://www.liveinternet.ru/stat/ru/searches.csv?period=week;graph=csv")

## ---

from yahoo_finance import Share
from yahoo_finance import Share
from pandas.io.json import json_normalize
import datetime

def extract_date(x):
    return datetime.datetime.strptime(x, '%Y-%m-%d').date()
    
yndx = Share('YNDX')
stocks =  yndx.get_historical(dates[0].strftime("%Y-%m-%d"), dates[-1].strftime("%Y-%m-%d"))
df = json_normalize(stocks)
df.set_index("Date", inplace=True)
df['Close'] = df['Close'].apply(float)

f = plt.figure(figsize=(13, 5), dpi=80)

plt.subplots_adjust(left=0.05, bottom=0.1, right=0.87, top=0.93, wspace=0.27, hspace=0.36)

df.index = pandas.DatetimeIndex(df.index.map(extract_date))

df['Close'].resample('W', how='mean').plot()
```
