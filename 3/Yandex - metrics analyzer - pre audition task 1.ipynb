# Построим модели ARMA / ARIMA

```python
i = 0

y= []
for i in xrange(len(df)):
    y.append(df[i]['Monthly'])
    q = y[i].describe()
    print q
    fig = plt.figure(figsize=(5,5))
    fig = y[i].hist()
```

```python
row = ['JB', 'p-value', 'skew', 'kurtosis']


jb_tests = []

for i in xrange(len(df)):
    
    jb_tests.append(sm.stats.stattools.jarque_bera(y[i]))

a = np.vstack(jb_tests)
itog = SimpleTable(a, row)
itog
```

```python
dickies = []

for i in xrange(len(df)):
    qx = sm.tsa.adfuller(y[i])
    
    if qx[0] > qx[4]['5%']:
        stationarity = 'не стационарный ряд, есть единичные корни'
    else: 
        stationarity = 'стационарный ряд, нет единичных корней'
        
    dickies.append([qx[0], qx[1], qx[4], stationarity])
    
row = ['adf', 'p-value', 'critical-values', 'stationarity']    

a = np.vstack(dickies)
itog = SimpleTable(a, row)
itog
```

```python
diffs = []

for i in xrange(len(df)):    
    diffs.append(y[i].diff().dropna())
    fig = plt.figure(figsize=(5,5))
    fig = diffs[i].plot()

dickies = []

for i in xrange(len(df)):
    qx = sm.tsa.adfuller(diffs[i])
    
    if qx[0] > qx[4]['5%']:
        stationarity = 'не стационарный ряд, есть единичные корни'
    else: 
        stationarity = 'стационарный ряд, нет единичных корней'
        
    dickies.append([qx[0], qx[1], qx[4], stationarity])
    
row = ['adf', 'p-value', 'critical-values', 'stationarity']    

a = np.vstack(dickies)
itog = SimpleTable(a, row)
itog    
```
